#<font color = 'gold'> Objetivo da utilização do Google Colab:</font>

A equipe viu a necessidade de aglomerar os dados de conexões do linkedin de todos os Participantes em um só DataSet para criar uma visualização no Power BI que demonstre da melhor forma esses dados.

<font color = 'gold'>Para alcançar esse objetivo foi necessário:</font>

1. Importar os dados de conexões do Linkedin de cada integrante;
2. Criar uma nova coluna **"Owner"** em cada tabela que guardasse a informação de quem era aquelas conexões;
3. Fazer a concatenação (Junção) de todas as tabelas já altaeradas com o nome de cada integrante.
4. Exportar o arquivo concatenado para uma unidade local e assim utilizá-lo para criar uma visualizção am alguma ferramenta adequada, no nosso caso, o Power BI

```python
import pandas as pd
import numpy as np
```

<Font color = 'gold'>Iportação de dados do linkedIn dos participantes do squad: </font>

```python
linkedinMatheus = pd.read_csv('Dados_M5 - LinkedIn_Matheus.csv')
#----------------------------------------------------------------------
linkedinCarol = pd.read_csv('Dados_M5 - Linkedin_Carol.csv')
#----------------------------------------------------------------------
linkedinCamilla = pd.read_csv('Dados_M5 - Linkedin_Camilla.csv')
#----------------------------------------------------------------------
linkedinElizangela = pd.read_csv('Dados_M5 - Linkedin_Elizangela.csv')
```

<font color = 'gold'> Criando uma nova coluna que guarda o nome do proprietário de cada tabela:

```python
linkedinMatheus['Owner'] = ['Matheus' for linha in linkedinMatheus['Company']]
#---------------------------------------------------------------------------------------
linkedinCarol['Owner'] = ['Carol' for linha in linkedinCarol['Company']]
#---------------------------------------------------------------------------------------
linkedinCamilla['Owner'] = ['Camilla' for linha in linkedinCamilla['Company']]
#---------------------------------------------------------------------------------------
linkedinElizangela['Owner'] = ['Elizangela' for linha in linkedinElizangela['Company']]
```

<font color = 'gold'> Concatenando todas as tabelas:</font>

```python
tabelasConcatenadas = pd.concat([linkedinMatheus,linkedinCarol,linkedinCamilla,linkedinElizangela])
```

<font color = 'gold'> Importação do método "Files" do google colab para que faça o download da tabela concatenada que foi demonstrada acima:</font>

```python
from google.colab import files

tabelasConcatenadas.to_csv('Conexoes_Squad.csv')

files.download('Conexoes_Squad.csv')
```
